# Diferenciação Numérica

Vamos aproximar derivadas em dados discretos com métodos simples.


Métodos desta aula:
- diferença progressiva;
- diferença regressiva;
- diferença central;
- `numpy.gradient()`.


```python
import numpy as np
import matplotlib.pyplot as plt
```

Exemplo: `f(x) = x²`.


```python
x = np.array([0, 1, 2, 3, 4, 5])   # pontos de x
y = x**2                            # valores de f(x)
h = x[1] - x[0]                     # distância entre pontos

print('x =', x)
print('y =', y)

plt.plot(x, y, 'o-', lw=2)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Função x²')
plt.grid(alpha=0.3)
plt.show()
```

Diferença progressiva usa o ponto atual e o próximo ponto.


```python
dy_prog = np.zeros_like(y, dtype=float)        # vetor da derivada
dy_prog[:-1] = (y[1:] - y[:-1]) / h           # fórmula progressiva
dy_prog[-1] = dy_prog[-2]                     # ajuste final

print('progressiva =', dy_prog)

plt.plot(x, dy_prog, 'o-', lw=2)
plt.xlabel('x')
plt.ylabel("f'(x)")
plt.title('Diferença progressiva')
plt.grid(alpha=0.3)
plt.show()
```

Diferença regressiva usa o ponto atual e o ponto anterior.


```python
dy_reg = np.zeros_like(y, dtype=float)         # vetor da derivada
dy_reg[1:] = (y[1:] - y[:-1]) / h             # fórmula regressiva
dy_reg[0] = dy_reg[1]                         # ajuste inicial

print('regressiva =', dy_reg)

plt.plot(x, dy_reg, 'o-', lw=2)
plt.xlabel('x')
plt.ylabel("f'(x)")
plt.title('Diferença regressiva')
plt.grid(alpha=0.3)
plt.show()
```

Diferença central usa um ponto antes e um depois.


```python
dy_cent = np.zeros_like(y, dtype=float)        # vetor da derivada
dy_cent[1:-1] = (y[2:] - y[:-2]) / (2*h)      # fórmula central
dy_cent[0] = (y[1] - y[0]) / h                # borda esquerda
dy_cent[-1] = (y[-1] - y[-2]) / h             # borda direita

print('central =', dy_cent)

plt.plot(x, dy_cent, 'o-', lw=2)
plt.xlabel('x')
plt.ylabel("f'(x)")
plt.title('Diferença central')
plt.grid(alpha=0.3)
plt.show()
```

O `numpy.gradient()` faz essa aproximação automaticamente em dados discretos.


```python
dy_grad = np.gradient(y, x)                    # derivada automática

print('gradient =', dy_grad)

plt.plot(x, dy_grad, 'o-', lw=2)
plt.xlabel('x')
plt.ylabel("f'(x)")
plt.title('numpy.gradient()')
plt.grid(alpha=0.3)
plt.show()
```

Resumo: cada método usa os pontos disponíveis de um jeito diferente para aproximar a derivada.
