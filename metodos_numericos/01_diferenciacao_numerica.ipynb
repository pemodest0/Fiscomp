# Diferenciação Numérica

Vamos aproximar a derivada de uma função conhecida e comparar os erros dos métodos.


Função usada nesta aula:
- `f(x) = x²`
- derivada exata: `f'(x) = 2x`


```python
import numpy as np
```

Escolhemos alguns pontos de `x` e calculamos os valores exatos da derivada.


```python
x = np.array([0, 1, 2, 3, 4, 5])            # pontos de x
y = x**2                                     # valores de f(x)
h = x[1] - x[0]                              # distância entre pontos
dy_exato = 2 * x                             # derivada exata
escala = np.mean(np.abs(dy_exato[1:]))       # base para porcentagem
erros = {}                                   # guarda os erros

print('f(x) = x^2')
print("f'(x) = 2x")
print('x =', x)
print('f(x) =', y)
print("f'(x) exata =", dy_exato)

def mostrar(nome, aprox):
    erro = np.mean(np.abs(aprox - dy_exato))  # erro médio
    print(f"\n{nome}")
    print('derivada numérica =', aprox)
    print(f'erro médio = {erro:.6e}')
    erros[nome] = erro
```

## Método 1: diferença progressiva


```python
dy_prog = np.zeros_like(y, dtype=float)       # vetor da derivada
dy_prog[:-1] = (y[1:] - y[:-1]) / h           # fórmula progressiva
dy_prog[-1] = dy_prog[-2]                     # ajuste final

mostrar('Progressiva', dy_prog)
```

## Método 2: diferença regressiva


```python
dy_reg = np.zeros_like(y, dtype=float)        # vetor da derivada
dy_reg[1:] = (y[1:] - y[:-1]) / h             # fórmula regressiva
dy_reg[0] = dy_reg[1]                         # ajuste inicial

mostrar('Regressiva', dy_reg)
```

## Método 3: diferença central


```python
dy_cent = np.zeros_like(y, dtype=float)       # vetor da derivada
dy_cent[1:-1] = (y[2:] - y[:-2]) / (2 * h)    # fórmula central
dy_cent[0] = (y[1] - y[0]) / h                # borda esquerda
dy_cent[-1] = (y[-1] - y[-2]) / h             # borda direita

mostrar('Central', dy_cent)
```

## Método 4: `numpy.gradient()`


```python
dy_grad = np.gradient(y, x)                   # derivada automática

mostrar('Gradient', dy_grad)
```

## Comparação percentual dos erros


```python
print('Erro percentual de cada método:')
for nome, erro in erros.items():
    perc = 100 * erro / escala
    print(f'{nome:12s}: {perc:.2f}%')
```
