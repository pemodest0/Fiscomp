# Integração Numérica

Vamos aproximar a mesma integral com métodos diferentes e comparar os erros.


Função usada nesta aula:
- `f(x) = x²`
- integral exata: `∫₀¹ x² dx = 1/3`


```python
import numpy as np
from scipy.integrate import quad, trapezoid, simpson
```

Escolhemos alguns pontos de `x` no intervalo `[0, 1]`.


```python
n = 6                                         # número de pontos
x = np.linspace(0, 1, n)                      # pontos de x
y = x**2                                      # valores de f(x)
h = x[1] - x[0]                               # distância entre pontos
area_exata = 1/3                              # valor exato
erros = {}                                    # guarda os erros

print('f(x) = x^2')
print('Integral exata: ∫_0^1 x^2 dx = 1/3')
print('x =', x)
print('f(x) =', y)

def mostrar(nome, valor):
    erro = abs(valor - area_exata)            # erro absoluto
    print(f"\n{nome}")
    print(f'valor numérico = {valor:.6f}')
    print(f'erro absoluto = {erro:.6e}')
    erros[nome] = erro
```

## Método 1: retângulos pela esquerda


```python
area_ret = 0                                  # começa em zero
for i in range(len(x) - 1):                   # percorre os intervalos
    area_ret += y[i] * h                      # soma os retângulos

mostrar('Retângulos', area_ret)
```

## Método 2: trapézios com `for`


```python
area_trap = 0                                 # começa em zero
for i in range(len(x) - 1):                   # percorre os intervalos
    area_trap += (y[i] + y[i + 1]) * h / 2    # soma os trapézios

mostrar('Trapézios', area_trap)
```

## Método 3: `quad`


```python
area_quad, erro_quad = quad(lambda z: z**2, 0, 1)

mostrar('quad', area_quad)
```

## Método 4: `trapezoid`


```python
area_trapezoid = trapezoid(y, x)

mostrar('trapezoid', area_trapezoid)
```

## Método 5: `simpson`


```python
area_simpson = simpson(y, x=x)

mostrar('simpson', area_simpson)
```

## Comparação percentual dos erros


```python
print('Erro percentual de cada método:')
for nome, erro in erros.items():
    perc = 100 * erro / abs(area_exata)
    print(f'{nome:12s}: {perc:.2f}%')
```
