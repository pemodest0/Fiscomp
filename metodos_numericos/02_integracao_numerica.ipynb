# Integração Numérica

Vamos ver integração de área e integração de EDO com exemplos curtos.


Métodos desta aula:
- retângulos com `for`;
- trapézios com `for`;
- `quad`;
- `trapezoid`;
- `simpson`;
- `odeint`.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad, trapezoid, simpson, odeint
```

Primeiro, vamos aproximar a área de `f(x) = x²` em `[0, 1]`.


```python
n = 6                            # número de pontos
x = np.linspace(0, 1, n)         # pontos de x
y = x**2                         # valores de f(x)
h = x[1] - x[0]                  # distância entre pontos

print('x =', x)
print('y =', y)
```

Retângulos pela esquerda, com `for`.


```python
area_ret = 0                     # começa em zero
for i in range(len(x) - 1):      # percorre os intervalos
    area_ret += y[i] * h         # soma o retângulo

print('retângulos =', area_ret)
```

Trapézios, também com `for`.


```python
area_trap = 0                                    # começa em zero
for i in range(len(x) - 1):                       # percorre os intervalos
    area_trap += (y[i] + y[i + 1]) * h / 2       # soma o trapézio

print('trapézios =', area_trap)
```

Agora usamos funções prontas do SciPy.


```python
res_quad, err_quad = quad(lambda z: np.sin(z), 0, np.pi/2)   # quadratura
area_trapezoid = trapezoid(y, x)                             # trapézios prontos
area_simpson = simpson(y, x=x)                               # Simpson

print('quad em sen(x)  =', res_quad)
print('trapezoid em x² =', area_trapezoid)
print('simpson em x²   =', area_simpson)
```

O gráfico ajuda a visualizar a área sob a curva.


```python
plt.plot(x, y, 'o-', lw=2, label='f(x) = x²')
plt.fill_between(x, y, alpha=0.25)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Área sob a curva')
plt.grid(alpha=0.3)
plt.show()
```

Também podemos integrar uma EDO no tempo. Exemplo: `y' = -2y`.


```python
def modelo(y, t):
    return -2 * y                              # EDO

y0 = 5                                         # condição inicial
t = np.linspace(0, 5, 100)                     # pontos no tempo
y = odeint(modelo, y0, t)                      # solução numérica

print('y(0) =', y[0, 0])
print('y(5) =', y[-1, 0])

plt.plot(t, y, lw=2)
plt.xlabel('tempo')
plt.ylabel('y(t)')
plt.title('Solução numérica da EDO')
plt.grid(alpha=0.3)
plt.show()
```
